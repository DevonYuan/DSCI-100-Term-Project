#### Task 5.2 Perform the Analysis (20 points)

- Perform the proposed analysis.
- Be sure to comment your code throughout so that it is easy to follow. 

*You may add more code cells if needed.* 


</div>

```r
# We will need to load this library
library(recipes)

# Create a new dataframe - I will the analysis 50 times 
k_values <- data.frame(matrix(ncol = 1, nrow = 0))
colnames(k_values) <- c("K")

# Run a for loop to conduct the analysis 50 times 
for (x in 1:50) {
    # First, split the data into the training and testing split 
    # Make sure to use stratified sampling on the tournament finish variable 
    basketball_split <- initial_split(tidy_basketball_data, prop = 0.75, strata = tourney_finish)
    basketball_training <- training(basketball_split)
    basketball_testing <- testing(basketball_split)

    # We will use one-hot encoding to use bid as a predictor variable 
    basketball_recipe <- recipe(tourney_finish ~ seed + reg_wins_pct + conf_wins_pct +  
                                tourney_wins_pct + bid + total_wins_pct, data = basketball_training) |>
    step_dummy(all_nominal_predictors(), one_hot = TRUE) |>
    step_novel(all_nominal_predictors()) |>
    step_zv(all_predictors()) |>
    step_scale(all_predictors()) |>
    step_center(all_predictors()) 

    # Create the KNN model 
    basketball_spec <- nearest_neighbor(weight_func = "rectangular", neighbors = tune()) |>
    set_engine("kknn") |>
    set_mode("classification")

    # Create the workflow 
    basketball_workflow <- workflow() |>
    add_recipe(basketball_recipe) |>
    add_model(basketball_spec) |> 
    fit(data = basketball_training)

    # Create the vfold 
    basketball_vfold <- vfold_cv(basketball_training, v = 5, strata = tourney_finish)

    # I will have to decide on choosing the optimal value of K, from 1 to 10 neighbors 
    gridvals <- tibble(neighbors = seq(from = 1, to = 10, by = 1))

    # Perform the tuning 
    basketball_results <- basketball_workflow |>
    tune_grid(resamples = basketball_vfold, grid = gridvals) |>
    collect_metrics() |>
    filter(.metric == "accuracy")

    # Fetch the value giving the highest accuracy 
    optimal_k_value <- basketball_results |>
    arrange(desc(mean)) |>
    select(neighbors) |>
    head(1) |>
    pull()

    # Create a row representign the optimal value of K created from this iteration of the analysis 
    this_row <- tibble(K = c(optimal_k_value))

    # Add it to the current dataframe 
    k_values <- bind_rows(k_values, this_row)
}
```

```r
# Now, we are going to complete the analysis by selecting the mode 
k_value <- k_values |>
table() |>
which.max() |>
names() |>
as.numeric()

# Now, we will perform the analysis after having determined our K value 
basketball_spec <- nearest_neighbor(weight_func = "rectangular", neighbors = k_value) |>
set_engine("kknn") |>
set_mode("classification")

# Remake the workflow with the new model 
basketball_fit <- workflow() |>
add_recipe(basketball_recipe) |>
add_model(basketball_spec) |>
fit(data = basketball_training)

# Test against testing set 
basketball_summary <- basketball_fit |>
predict(basketball_testing) |> 
bind_cols(basketball_testing) |>
metrics(truth = tourney_finish, estimate = .pred_class) |>
filter(.metric == "accuracy")

k_value
basketball_summary
```
